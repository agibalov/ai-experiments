```python
%matplotlib inline

import pandas as pd

data = pd.read_csv("kc_house_data.csv")
data.head()
```

```python
import seaborn as sns
sns.distplot(data["price"])
```

```python
sns.jointplot(x=data["price"], y=data["sqft_living"])
```

```python
sns.pairplot(data, vars=["price", "sqft_living", "sqft_lot", "sqft_basement"], hue="floors")
```

```python
c = data.corr()
sns.heatmap(c, cmap=sns.diverging_palette(220, 10, as_cmap=True), square=True)
```
