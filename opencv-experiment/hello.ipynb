```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('1.jpg')
(height, width, depth) = image.shape
print(f'{width}x{height}x{depth}')
```

```python
import numpy as np

experiments = {
    'original': image,
    'crop': image[100:500, 200:600],
    'resize': cv2.resize(image, (64, 48)),
    'gaussian blur': cv2.GaussianBlur(image, (41, 41), 0),
    'median blur': cv2.medianBlur(image, 41),
    'bilateral filter': cv2.bilateralFilter(image[100:500, 200:600], 9, 75, 75),
    'erode': cv2.erode(image, np.ones((5, 5), np.uint8), iterations=3),
    'dilate': cv2.dilate(image, np.ones((5, 5), np.uint8), iterations=1),
    'morphological gradient': cv2.morphologyEx(
        image[100:500, 200:600], cv2.MORPH_GRADIENT, np.ones((5, 5), np.uint8)),
    'edges': cv2.Canny(cv2.cvtColor(image[100:500, 200:600], cv2.COLOR_BGR2GRAY), 100, 255)
}

plt.figure(figsize=(12, 16))

for i, (title, im) in enumerate(experiments.items()):
    plt.subplot(4, 3, i + 1)
    plt.title(title)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(cv2.cvtColor(im, cv2.COLOR_BGR2RGB))
```
