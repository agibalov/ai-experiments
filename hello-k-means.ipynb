```python
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

num_of_blobs = 3
x, blob_index = make_blobs(1000, centers=num_of_blobs, random_state=1)
cluster_index4 = KMeans(n_clusters=4, random_state=2).fit_predict(x)
cluster_index3 = KMeans(n_clusters=3, random_state=2).fit_predict(x)
cluster_index2 = KMeans(n_clusters=2, random_state=2).fit_predict(x)

plt.figure(figsize=(12, 12))

plt.subplot(2, 2, 1)
plt.title("Original %d blobs" % (num_of_blobs,))
plt.scatter(x[:, 0], x[:, 1], c=blob_index)

plt.subplot(2, 2, 2)
plt.title("K-means with clusters=2")
plt.scatter(x[:, 0], x[:, 1], c=cluster_index2)

plt.subplot(2, 2, 3)
plt.title("K-means with clusters=3")
plt.scatter(x[:, 0], x[:, 1], c=cluster_index3)

plt.subplot(2, 2, 4)
plt.title("K-means with clusters=4")
plt.scatter(x[:, 0], x[:, 1], c=cluster_index4)

plt.show()
```
