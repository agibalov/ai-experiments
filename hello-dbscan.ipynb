```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 12))

for i in range(1, 5):
    num_of_blobs = i + 1
    x, blob_index = make_blobs(1000, centers=num_of_blobs, random_state=1)
    db = DBSCAN().fit(x)
    #db = DBSCAN(min_samples = 10).fit(x)
    #db = DBSCAN(eps = 0.3).fit(x)
    unique_labels = set(db.labels_)
    num_of_clusters = len(unique_labels)

    plt.subplot(2, 2, i)
    plt.title("Generated: %d, detected: %d" % (num_of_blobs, num_of_clusters,))
    plt.scatter(x[:, 0], x[:, 1], c=db.labels_, cmap=plt.cm.rainbow)

plt.show()
```
